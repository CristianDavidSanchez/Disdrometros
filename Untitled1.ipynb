```python
import pandas as pd
import glob

# Obtén la lista de archivos CSV en la carpeta
archivos_csv = glob.glob('C:/Users/pc/EAFITDATA/Solicitud_Disdros2/*.csv')

# Crea una lista vacía para almacenar los DataFrames
dataframes = []

# Itera sobre cada archivo CSV y conviértelo en un DataFrame
for archivo in archivos_csv:
    df = pd.read_csv(archivo)
    df= df.dropna()
    df = df.loc[df['var6'] != 0]
    dfzeros = df
    columnas_eliminar = ['cliente', 'Unnamed: 0','var6','calidad','var8','var9','var10','var11','var13']
    dfzeros = dfzeros.drop(columnas_eliminar, axis=1)
    indices_ceros = dfzeros.loc[(dfzeros == 0).all(axis=1)].index
    df = df.drop(indices_ceros)
    dataframes.append(df)
# Concatena los DataFrames en uno solo
dataframe_final = pd.concat(dataframes, ignore_index=True)
dataframe_final
```

```python
dataframe_final.describe()
```

```python
import matplotlib.pyplot as plt
df2=dataframe_final.groupby(by="var6")
conteo_por_grupo=df2.size()
plt.bar(conteo_por_grupo.index, conteo_por_grupo.values)
plt.xlabel('Clasificación')
plt.ylabel('Datos')
plt.title('Datos por clasificación')
plt.show()
print(conteo_por_grupo)
```
